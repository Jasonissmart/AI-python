```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from matplotlib.pylab import date2num
data1=pd.read_csv("@EC-分鐘-成交價.csv",index_col=0,sep=",",engine='python')
data1.index=pd.to_datetime(data1.index)
data1=pd.DataFrame(data1)
data1.head()

#new_data=new_data.loc['2001/5/13':'2015/12/30']
```

```python
import matplotlib
data_Open=data1.loc[:, 'Open']
data_High=data1.loc[:, 'High']
data_Low=data1.loc[:, 'Low']
data_Close=data1.loc[:, 'Close']
spd1=data_High-data_Low
plot(spd1)
```

```python
data_Open=data1.loc[:, 'Open'].resample('60T').first().dropna()
data_High=data1.loc[:, 'High'].resample('60T').max().dropna()
data_Low=data1.loc[:, 'Low'].resample('60T').min().dropna()
data_Close=data1.loc[:, 'Close'].resample('60T').last().dropna()
#data_TotalVolumn=data1.loc[:,'TotalVolume'].resample('60T').sum().dropna()

new_data=pd.concat([data_Open,data_High,data_Low,data_Close],axis=1)
print(new_data)
```

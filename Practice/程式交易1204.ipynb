```python
# -*- coding: UTF-8 -*-
#設定買賣時間(日期),停損點數
#取得歷史資料
Match_data = [ line.strip('\n').split(",") for line in open('TXFB8.I020') ][1:]

Match_data
```

```python
#取得歷史日期
Date_list = set([ i[0] for i in Match_data ])
#Date_list.remove()
Date_list=sorted(list(Date_list))
```

```python
#每日回測
for DateNum in range(len(Date_list)):
    #日期範圍
    #if int(Date_list[DateNum])>=20160126 and int(Date_list[DateNum])<=20160216:

        #取得回測每日資料
        C_I020 = [ i for i in Match_data if i[0]== Date_list[DateNum] ]
        
        #設立買賣時間,預設開盤買收盤賣
        I020a= [ line for line in C_I020 if int(line[1])>8450000 and int(line[1])<13460000 ]
        
        OrderTime=I020a[0][1]  #下單時間紀錄
        OrderPrice=int(I020a[0][3]) #下單價格紀錄


        for i in range(1,len(I020a)):
            price=int(I020a[i][3])
            if price <= OrderPrice-50 or price >= OrderPrice+50:
                CoverTime=I020a[i][1]  #平倉時間紀錄
                CoverPrice=int(I020a[i][3])  #平倉時間紀錄
                break
            elif i == len(I020a)-1:
                CoverTime=I020a[i][1]  #平倉時間紀錄
                CoverPrice=int(I020a[i][3]) #平倉時間紀錄
  
        print("Date:",Date_list[DateNum]," OrderTime:",OrderTime," OrderPrice:",OrderPrice," CoverTime:",CoverTime," CoverPrice:",CoverPrice," Profit:",CoverPrice-OrderPrice)
```
